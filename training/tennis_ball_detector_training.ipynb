# 0. Downloading the Roboflow Dataset

```python
from roboflow import Roboflow
rf = Roboflow(api_key="API_KEY")
project = rf.workspace("test-rhs4w").project("tennis-ball-s4gp5")
version = project.version(1)
dataset = version.download("yolov11")
```

# 1. Config

```python
import os, yaml

yaml_in = f"{dataset.location}/data.yaml"
inner_root = os.path.join(dataset.location, "tennis-ball-1")

# sanity check
assert os.path.isdir(os.path.join(inner_root, "train", "images")), "train/images not found in inner_root"
assert os.path.isdir(os.path.join(inner_root, "valid", "images")), "valid/images not found in inner_root"

with open(yaml_in, "r") as f:
    d = yaml.safe_load(f)

# Patch to correct root + correct relative paths
d["path"] = inner_root
d["train"] = "train/images"
d["val"] = "valid/images"
d["test"] = "test/images"

yaml_out = f"{dataset.location}/data_fixed.yaml"
with open(yaml_out, "w") as f:
    yaml.safe_dump(d, f, sort_keys=False)

print("✅ Wrote:", yaml_out)
print("✅ path:", d["path"])
print("✅ train/val/test:", d["train"], d["val"], d["test"])
```
