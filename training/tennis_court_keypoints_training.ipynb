# Tennis Court Keypoints Training

## 1. Imports

```python
import os
import json

import cv2
import numpy as np
import torch
from torch import nn
from torch.utils.data import DataLoader, Dataset
from torchvision import models, transforms
```

## 2. Config

```python
# Config (works whether notebook is run from project root or training/)
images_dir = "data/images"
train_json = "data/data_train.json"
val_json = "data/data_val.json"
output = "../models/keypoints_model.pth"

if not os.path.exists(images_dir):
    images_dir = "training/data/images"
    train_json = "training/data/data_train.json"
    val_json = "training/data/data_val.json"
    output = "models/keypoints_model.pth"

epochs = 30
batch_size = 32
image_size = 224
lr = 3e-4
weight_decay = 1e-4
workers = 4
patience = 8
```

## 3. PyTorch Dataset

```python
class KeypointsDataset(Dataset):
    def __init__(self, img_dir, data_file, image_size=224, augment=False):
        self.img_dir = img_dir
        self.image_size = image_size
        self.augment = augment

        with open(data_file, "r") as f:
            self.data = json.load(f)

        self.to_pil = transforms.ToPILImage()
        self.base_transforms = transforms.Compose(
            [
                transforms.Resize((image_size, image_size)),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            ]
        )
        self.augment_transforms = transforms.Compose(
            [
                transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.15, hue=0.02),
            ]
        )

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        image_path = os.path.join(self.img_dir, f"{item['id']}.png")
        img = cv2.imread(image_path)
        if img is None:
            raise FileNotFoundError(f"Failed to read image: {image_path}")

        h, w = img.shape[:2]
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = self.to_pil(img)

        if self.augment:
            img = self.augment_transforms(img)
        img = self.base_transforms(img)

        keypoints = np.array(item["kps"], dtype=np.float32).reshape(-1)
        keypoints[::2] *= float(self.image_size) / float(w)
        keypoints[1::2] *= float(self.image_size) / float(h)
        keypoints = torch.from_numpy(keypoints)
        return img, keypoints
```

## 4. ResNet50 model

```python
def build_model(num_keypoints=14):
    try:
        model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
    except AttributeError:
        model = models.resnet50(pretrained=True)
    model.fc = nn.Linear(model.fc.in_features, num_keypoints * 2)
    return model


def run_epoch(model, loader, optimizer, criterion, device, training):
    if training:
        model.train()
    else:
        model.eval()

    running_loss = 0.0
    num_samples = 0

    with torch.set_grad_enabled(training):
        for images, keypoints in loader:
            images = images.to(device, non_blocking=True)
            keypoints = keypoints.to(device, non_blocking=True)

            outputs = model(images)
            loss = criterion(outputs, keypoints)

            if training:
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

            batch_size = images.shape[0]
            running_loss += loss.item() * batch_size
            num_samples += batch_size

    return running_loss / max(num_samples, 1)
```

## 5. Training

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

train_dataset = KeypointsDataset(
    img_dir=images_dir,
    data_file=train_json,
    image_size=image_size,
    augment=True,
)
val_dataset = KeypointsDataset(
    img_dir=images_dir,
    data_file=val_json,
    image_size=image_size,
    augment=False,
)

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=workers,
    pin_memory=torch.cuda.is_available(),
)
val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=workers,
    pin_memory=torch.cuda.is_available(),
)

model = build_model().to(device)
criterion = nn.SmoothL1Loss(beta=1.0)
optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode="min", factor=0.5, patience=3, min_lr=1e-6
)

output_dir = os.path.dirname(output)
if output_dir:
    os.makedirs(output_dir, exist_ok=True)

best_path = output.replace(".pth", "_best.pth")
last_path = output.replace(".pth", "_last.pth")

best_val_loss = float("inf")
epochs_without_improvement = 0

for epoch in range(epochs):
    train_loss = run_epoch(model, train_loader, optimizer, criterion, device, training=True)
    val_loss = run_epoch(model, val_loader, optimizer, criterion, device, training=False)
    scheduler.step(val_loss)

    current_lr = optimizer.param_groups[0]["lr"]
    print(
        f"Epoch {epoch + 1:03d}/{epochs:03d} | "
        f"train_loss={train_loss:.4f} | val_loss={val_loss:.4f} | lr={current_lr:.6f}"
    )

    torch.save(model.state_dict(), last_path)

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_without_improvement = 0
        torch.save(model.state_dict(), best_path)
        torch.save(model.state_dict(), output)
        print(f"  -> Saved new best model to {output}")
    else:
        epochs_without_improvement += 1
        if epochs_without_improvement >= patience:
            print(f"Early stopping at epoch {epoch + 1}. Best val_loss={best_val_loss:.4f}")
            break

print(f"Training complete. Best val_loss={best_val_loss:.4f}")
print(f"Best checkpoint: {best_path}")
print(f"Last checkpoint: {last_path}")
```
